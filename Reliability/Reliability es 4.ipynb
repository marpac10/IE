```python
def esponenziale (x,l) :
    return np.exp(-l*x)

t = np.linspace(0,5000,1000)

s11= 1-(1-np.exp(-t/5000)*np.exp(-t/1000))*(1-np.exp(-t/2000)*np.exp(-t/1000))
s12 =np.exp(-t/1000)*(1-(1-np.exp(-t/5000))*(1-np.exp(-t/2000)))
 
plt.figure(figsize = (20, 10))
plt.plot(t,s11,label='Sistema 1')
plt.plot(t,s12,label='Sistema 2',color='red')
plt.legend(loc='upper right',fontsize = 'x-large')
plt.title('Confronto sistemi S1 - S2', size=22)
plt.xlabel('Tempo', size=16)
plt.ylabel('Reliability', size=16)
plt.grid()
plt.show()

```
