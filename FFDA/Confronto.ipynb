```python
import pandas as pd
import numpy as np
import sys
import seaborn as sn
import io
import os
import pathlib
import matplotlib.pyplot as plt
import reliability
import dataframe_image as dfi
from datetime import datetime as dt
import itertools
from collections import Counter


# path_tool = str(pathlib.Path().absolute())+'/ffdatoolsetIE2018/'
path = str(pathlib.Path().absolute())
```

```python
mercury_int = pd.read_csv(path+'/mercury.csv')
bluegene_int = pd.read_csv(path+'/bluegene.csv')
```

```python
mercury_x = np.sort(mercury_int.Interarrivals)
mercury_x = [i for i in  mercury_x if i <= 200000]
mercury_y = np.arange(len(mercury_x))/float(len(mercury_x))

bg_x = np.sort(bluegene_int.Interarrivals)
bg_x = [i for i in  bg_x if i <= 200000]
bg_y = np.arange(len(bg_x))/float(len(bg_x))

plt.figure(figsize = (10, 6))
# plt.plot(mercury_x, mercury_y, label="Empirical CDF")
plt.plot(mercury_x,1-mercury_y, label="Reliability Mercury")
plt.plot(bg_x,1-bg_y, label="Reliability BlueGene/L")
plt.title('Reliability', size=22)
plt.xlabel('Tempo (s)', size=16)
plt.ylabel('Probabilità', size=16)
plt.legend(loc='upper right',fontsize = 'x-large')
plt.grid()
plt.show()
```
